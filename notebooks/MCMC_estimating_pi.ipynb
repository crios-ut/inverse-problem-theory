```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Provide number of points N
N = 1000

# get random coordinates (x,y) in the unit square [0,1] x [0,1]
x = np.random.rand(N)
y = np.random.rand(N)

# calculate distance between (0,0) and (x,y)
r = x*x + y*y

# count number of points inside unit circle centered at 0
hit = r[r<=1]

myPi = 4*len(hit)/N
relErr = abs(np.pi-myPi)/np.pi


print("Using {} points, approximation to pi is {:.4f}".format(N, myPi))
print("Relative error = {:.3f}%".format(100*relErr))
```

```python
# Plot all points
plt.scatter(x,y,c='b')

# Plot just points inside quarter circle in a different color
xHit = x[r<=1]
yHit = y[r<=1]

plt.scatter(x,y)
plt.scatter(xHit,yHit,c='g')
 
    
# plot circle
angle = np.linspace( 0 , np.pi/2 , 150 ) 
xCirc = np.cos( angle ) 
yCirc = np.sin( angle ) 
plt.plot(xCirc,yCirc,'r')
```
